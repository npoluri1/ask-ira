## NumPy, Pandas, and Data Analysis Basics

```python
import numpy as np
import pandas as pd

print(f"NumPy {np.__version__}, Pandas {pd.__version__}")
```

## NumPy Basics

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(f"Shape: {arr.shape}, Dims: {arr.ndim}")
print(f"Mean: {arr.mean():.2f}, Std: {arr.std():.2f}")
print(f"Broadcast: {arr * 2}")
```

## Pandas for Financial Data

```python
dates = pd.date_range('2024-01-01', periods=5, freq='D')
prices = pd.DataFrame({
    'AAPL': [180, 182, 178, 185, 190],
    'MSFT': [380, 385, 378, 390, 395],
    'GOOGL': [140, 142, 138, 145, 148],
}, index=dates)

returns = prices.pct_change()
print("Daily Returns:")
print(returns.round(4))
print(f"\nCorrelation:\n{prices.corr().round(3)}")
```

```python
# Vectorized operations for financial metrics
sharpe = returns.mean() / returns.std() * np.sqrt(252)
print(f"Sharpe Ratios (annualized):\n{sharpe.round(2)}")
```
