## 0 ¬∑ Basic Types

Python has four fundamental built-in scalar types:

| Type    | Keyword  | Example literals          |
|---------|----------|---------------------------|
| Integer | `int`    | `0`, `42`, `-7`, `2002`   |
| Float   | `float`  | `1.75`, `3.14`, `-0.5`    |
| Boolean | `bool`   | `True`, `False`           |
| Nothing | `None`   | `None`                    |

All other types (strings, lists, dicts ‚Ä¶) are built on top of these.

#### 0a ¬∑ Numeric types ‚Äî `int` and `float`

`int` holds whole numbers (no size limit in Python); `float` holds decimal values using IEEE 754 double precision.

````python
# ‚îÄ‚îÄ 0a ¬∑ int and float ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
birth_year  = 2002          # int  ‚Äî whole number
height_m    = 1.75          # float ‚Äî decimal value
salary      = 85_000        # int  ‚Äî underscores allowed for readability

print(f"birth_year : {birth_year}   type ‚Üí {type(birth_year)}")
print(f"height_m   : {height_m}    type ‚Üí {type(height_m)}")
print(f"salary     : {salary}  type ‚Üí {type(salary)}")

# Arithmetic
age        = 2026 - birth_year
bmi        = 70 / (height_m ** 2)   # ** is the power operator
print(f"\nAge  : {age}")
print(f"BMI  : {bmi:.2f}")          # .2f rounds to 2 decimal places

# int / float interaction
print(f"\n10 / 3  = {10 / 3}")      # true division ‚Üí float
print(f"10 // 3 = {10 // 3}")       # floor division ‚Üí int
print(f"10 % 3  = {10 % 3}")        # modulo (remainder)
````

#### 0b ¬∑ Boolean ‚Äî `bool`

`bool` is a subtype of `int` with exactly two values: `True` (== 1) and `False` (== 0).  
Comparison operators (`==`, `!=`, `<`, `>`, `in`, `not in`) all return a `bool`.

````python
# ‚îÄ‚îÄ 0b ¬∑ bool ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
is_active   = True
is_verified = False

print(f"is_active   : {is_active}   type ‚Üí {type(is_active)}")
print(f"is_verified : {is_verified}  type ‚Üí {type(is_verified)}")

# Comparisons produce booleans
age = 2026 - 2002
print(f"\nage >= 18  : {age >= 18}")
print(f"age == 21  : {age == 21}")

# bool is a subtype of int ‚Äî True == 1, False == 0
print(f"\nTrue  + True  = {True + True}")    # 2
print(f"False + 1     = {False + 1}")        # 1
print(f"isinstance(True, int) = {isinstance(True, int)}")   # True

# Logical operators
print(f"\nis_active and is_verified : {is_active and is_verified}")
print(f"is_active or  is_verified : {is_active or  is_verified}")
print(f"not is_active             : {not is_active}")
````

#### 0c ¬∑ `None` ‚Äî the absence of a value

`None` is Python's null sentinel. It is its own type (`NoneType`) and is the default return value of functions that don't explicitly return anything.

````python
# ‚îÄ‚îÄ 0c ¬∑ None ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
middle_name = None    # Hanu has no middle name on record

print(f"middle_name : {middle_name}   type ‚Üí {type(middle_name)}")

# Always test for None with 'is', not '=='
if middle_name is None:
    print("No middle name recorded.")

# .get() on a dict returns None when the key is missing
dept = hanu.get("department")   # key doesn't exist
print(f"\nhanu.get('department') ‚Üí {dept}")
print(f"Is None? {dept is None}")

# None as a default sentinel in a function signature
def greet(name, title=None):
    if title is None:
        return f"Hello, {name}!"
    return f"Hello, {title} {name}!"

print(greet("Hanu"))
print(greet("Shankar", title="Mr"))
````

## 1 ¬∑ Strings

Strings are immutable sequences of characters.  
Key operations: `+` concatenation, f-strings, `.split()`, `.strip()`, `.lower()`, slicing.

#### 1a ¬∑ String concatenation ‚Äî `+`, f-strings, `.join()`

These are the core composition patterns for building strings from multiple parts.

````python
# ‚îÄ‚îÄ 1a ¬∑ Building a full name ‚Äî concatenation, f-strings, join ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
fn, ln = "Hanu", "Madala"

by_plus  = fn + " " + ln
by_fstr  = f"{fn} {ln}"
by_join  = " ".join([fn, ln])

print("+ concat :", by_plus)
print("f-string :", by_fstr)
print(".join()  :", by_join)
````

#### 1b ¬∑ String transformation methods ‚Äî `.upper()`, `.split()`, `.strip()`, `.lower()`

These methods normalize case, remove surrounding whitespace, and tokenize text.

````python
email = hanu["email"]           # "hanu.m@tdfs.com"

print(email.upper())            # HANU.M@TDFS.COM
print(email.lower())            # hanu.m@tdfs.com  (already lowercase)
print(email.split("@"))         # ['hanu.m', 'tdfs.com']

username, domain = email.split("@")
print(f"Username : {username}")
print(f"Domain   : {domain}")

# strip() removes leading/trailing whitespace
padded = "  Shankar Cherukuri  "
print(repr(padded.strip()))     # 'Shankar Cherukuri'
````

#### 1c ¬∑ String normalization ‚Äî `.replace()` and `len()`

Normalization converts variant formats into a consistent representation for validation.

````python
# Hanu's phone: "+65 1234 5678"  ‚Üí  strip country code and spaces
phone = hanu["phone"]
local = phone.replace("+65", "").replace(" ", "")
print(f"Raw phone    : {phone!r}")
print(f"Local digits : {local!r}  ({len(local)} digits)")
print(f"Is 8 digits? : {len(local) == 8}")

# Shankar's phone is already a plain 8-digit string
print(f"\nShankar raw  : {shankar['phone']!r}  ({len(shankar['phone'])} chars)")
````

#### 1d ¬∑ String slicing syntax ‚Äî `[start:stop:step]`

Slicing extracts substrings by position and supports stepping and reversal.

````python
ln = "Madala"

print(f"Full last name          : {ln}")
print(f"First 3 chars  [: 3]   : {ln[:3]}")      # Mad
print(f"From index 3   [3:]    : {ln[3:]}")      # ala
print(f"Middle chars   [1:4]   : {ln[1:4]}")     # ada
print(f"Every 2nd char [::2]   : {ln[::2]}")     # Mdl
print(f"Reversed       [::-1]  : {ln[::-1]}")    # aladam

# Works on any string ‚Äî including a full name
full = "Hanu Madala"
print(f"\nFirst 4 chars of full name: {full[:4]}")   # Hanu
````

## 2 ¬∑ Lists

Lists are **ordered** and **mutable** ‚Äî the workhorse collection in Python.

#### 2a ¬∑ List indexing, slicing, and mutation

Lists provide ordered access and in-place updates through indexing and mutating methods.

````python
# ‚îÄ‚îÄ 2a ¬∑ Indexing, slicing, and mutation ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
names = ["Hanu Madala", "Shankar Cherukuri"]

print("First  [0]  :", names[0])
print("Last  [-1]  :", names[-1])
print("Slice [0:1] :", names[0:1])

# append / remove
names.append("Test User")
print("After append:", names)
names.remove("Test User")
print("After remove:", names)
````

#### 2b ¬∑ List comprehensions ‚Äî projection syntax

Comprehensions create new lists concisely by projecting each element from an iterable.

````python
# Syntax:  [expression  for item in iterable]
# Extract every email from the people list
emails = [p["email"] for p in people]
print("All emails:", emails)

# Build full names
full_names = [f"{p['first_name']} {p['last_name']}" for p in people]
print("Full names:", full_names)

# Access a specific field for each person
phones = [p["phone"] for p in people]
print("Phones    :", phones)
````

#### 2c ¬∑ Filtered list comprehensions ‚Äî conditional projection

A trailing `if` condition in a comprehension filters elements during list construction.

````python
emails = [p["email"] for p in people]

# Keep only tdfs.com addresses
tdfs_emails = [e for e in emails if e.endswith("tdfs.com")]
print("TDFS emails  :", tdfs_emails)

# Keep only persons with international phones
intl_people = [p["first_name"] for p in people if p["phone"].startswith("+")]
print("Intl phones  :", intl_people)

# Combine transform + filter: uppercase name for local-phone holders only
local_upper = [
    f"{p['first_name']} {p['last_name']}".upper()
    for p in people
    if not p["phone"].startswith("+")
]
print("Local (upper):", local_upper)
````

## 3 ¬∑ Tuples

Tuples are **ordered** and **immutable** ‚Äî ideal for fixed records like a date of birth.  
Because they cannot be changed, they can be used as dictionary keys and set members.

#### 3a ¬∑ Tuple literals and index access

Tuples represent fixed-length ordered data and support positional access like lists.

````python
# ‚îÄ‚îÄ 3a ¬∑ Creating and accessing tuples ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
hanu_dob    = (20, "Jan", 2002)
shankar_dob = (30, "Mar", 1997)

print("Hanu's DOB      :", hanu_dob)
print("Shankar's DOB   :", shankar_dob)
print("Hanu birth year :", hanu_dob[2])   # index access
print("Length          :", len(hanu_dob))
````

#### 3b ¬∑ Tuple unpacking patterns

Unpacking binds tuple elements to variables in a single assignment statement.

````python
hanu_dob = (20, "Jan", 2002)

# Assign each element to its own variable in one line
day, month, year = hanu_dob
print(f"Day={day}  Month={month}  Year={year}")

# Extended unpacking with *rest
first, *rest = hanu_dob
print(f"First element : {first}")
print(f"Rest          : {rest}")
````

#### 3c ¬∑ Tuple immutability and `TypeError`

Tuples are immutable; assignment to an element raises `TypeError`.

````python
hanu_dob = (20, "Jan", 2002)

# Trying to mutate a tuple raises TypeError
try:
    hanu_dob[0] = 21         # attempt to change the day
except TypeError as e:
    print("TypeError:", e)

# Compare: a list version IS mutable
hanu_dob_list = list(hanu_dob)
hanu_dob_list[0] = 21
print("Mutated list:", hanu_dob_list)   # works fine
print("Original tuple unchanged:", hanu_dob)
````

#### 3d ¬∑ Tuple hashability as dictionary keys

Immutable tuples are hashable, so they can be used as dictionary keys.

````python
hanu_dob    = (20, "Jan", 2002)
shankar_dob = (30, "Mar", 1997)

# Tuples are hashable ‚Üí can be used as dict keys; lists cannot
dob_lookup = {
    hanu_dob:    "Hanu Madala",
    shankar_dob: "Shankar Cherukuri",
}
print(dob_lookup[(30, "Mar", 1997)])   # Shankar Cherukuri

# Attempting to use a list as a key raises TypeError
try:
    bad = {[20, "Jan", 2002]: "Hanu"}
except TypeError as e:
    print("TypeError:", e)
````

## 4 ¬∑ Sets

Sets are **unordered collections of unique values** ‚Äî great for deduplication and membership tests.

#### 4a ¬∑ Set comprehensions and deduplication

Sets keep unique values, making them ideal for deduplication and distinct extraction.

````python
# ‚îÄ‚îÄ 4a ¬∑ Set comprehension and deduplication ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Sets automatically discard duplicates
domains = {p["email"].split("@")[1] for p in people}
print("Unique domains :", domains)          # {'tdfs.com'}
print("Count          :", len(domains))    # 1 ‚Äî both records share the same domain

# Plain set literal ‚Äî manual deduplication
values = {"tdfs.com", "acme.com", "tdfs.com", "acme.com"}
print("Deduped        :", values)           # 2 unique items
````

#### 4b ¬∑ Set algebra ‚Äî union `|`, intersection `&`, difference `-`

Set operators compute combined, shared, and exclusive elements between sets.

````python
tdfs_emails  = {p["email"] for p in people}
other_emails = {"someone@acme.com", "hanu.m@tdfs.com"}   # hanu appears in both

print("tdfs_emails  :", tdfs_emails)
print("other_emails :", other_emails)
print()
print("Union        :", tdfs_emails | other_emails)   # all unique emails
print("Intersection :", tdfs_emails & other_emails)   # hanu.m@tdfs.com
print("Difference   :", tdfs_emails - other_emails)   # shankar only (in tdfs, not other)
````

#### 4c ¬∑ Set membership lookup ‚Äî `in`

Membership tests on sets are optimized for fast lookup of known values.

````python
known_emails = {p["email"] for p in people}

# in / not in ‚Äî constant-time regardless of set size
print("hanu.m@tdfs.com known?      ", "hanu.m@tdfs.com"   in known_emails)   # True
print("unknown@other.com known?    ", "unknown@other.com" in known_emails)   # False

# Practical use: check before processing
incoming = "shankar.c@tdfs.com"
if incoming in known_emails:
    print(f"{incoming} is a recognised address ‚Äî proceed.")
else:
    print(f"{incoming} is unknown ‚Äî reject.")
````

## 5 ¬∑ Dictionaries

Dictionaries are **key ‚Üí value** mappings, ordered by insertion (Python 3.7+).  
They are the natural representation for structured records like a person profile.

#### 5a ¬∑ Dictionary key access ‚Äî `[]` and `.get()`

Use `[]` for required keys and `.get()` for optional keys with defaults.

````python
# ‚îÄ‚îÄ 5a ¬∑ Accessing values ‚Äî `[]` and `.get()` ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
print("Hanu's last name :", hanu["last_name"])
print("Shankar's email  :", shankar["email"])

# .get() returns a default instead of raising KeyError for missing keys
print("Hanu's dept      :", hanu.get("department", "N/A"))   # key doesn't exist

# KeyError when using [] on a missing key
try:
    _ = hanu["department"]
except KeyError as e:
    print("KeyError:", e)
````

#### 5b ¬∑ Dictionary iteration interfaces ‚Äî `.keys()`, `.values()`, `.items()`

Dictionary views expose keys, values, and key-value pairs for iteration.

````python
print("‚îÄ‚îÄ .keys() ‚îÄ‚îÄ")
for key in hanu.keys():
    print(f"  {key}")

print("\n‚îÄ‚îÄ .values() ‚îÄ‚îÄ")
for value in hanu.values():
    print(f"  {value}")

print("\n‚îÄ‚îÄ .items() ‚îÄ (key, value) pairs ‚îÄ‚îÄ")
for key, value in shankar.items():
    print(f"  {key:12} ‚Üí {value}")
````

#### 5c ¬∑ Nested dictionary structures

Nested mappings model hierarchical records accessed through chained keys.

````python
# Wrap both records inside a parent "directory" dict
directory = {
    "hanu":    hanu,
    "shankar": shankar,
}

print("Top-level keys :", list(directory.keys()))

# Access nested field: directory ‚Üí person ‚Üí field
print("Hanu's email   :", directory["hanu"]["email"])
print("Shankar's phone:", directory["shankar"]["phone"])

# Iterate over the directory
for name, record in directory.items():
    print(f"  {name}: {record['first_name']} {record['last_name']}")
````

#### 5d ¬∑ Dictionary mutation operations ‚Äî add, update, delete

Mutation operations modify mappings in place via assignment, `del`, and `.pop()`.

````python
hanu_copy = dict(hanu)   # shallow copy ‚Äî don't mutate the original

# Add a new key
hanu_copy["department"] = "Engineering"
print("After add:", hanu_copy)

# Update an existing key
hanu_copy["department"] = "Data & AI"
print("After update:", hanu_copy["department"])

# Remove a key
del hanu_copy["department"]
print("After del:", hanu_copy.get("department", "gone"))

# .pop() removes and returns the value
email = hanu_copy.pop("email", None)
print("Popped email:", email)
````

## 6 ¬∑ JSON

JSON is the standard format for exchanging data between systems.  
Python's `json` module converts between Python objects and JSON strings/files with two key functions:

| Function | Direction |
|---|---|
| `json.dumps(obj)` | Python ‚Üí JSON string |
| `json.loads(s)` | JSON string ‚Üí Python |
| `json.dump(obj, f)` | Python ‚Üí JSON file |
| `json.load(f)` | JSON file ‚Üí Python |

#### 6a ¬∑ JSON string serialization ‚Äî `dumps()` and `loads()`

`dumps()` encodes Python objects to JSON text and `loads()` parses JSON text back to Python objects.

````python
import json

# Serialize to a pretty-printed JSON string
json_str = json.dumps(people, indent=2)
print(json_str)

# Deserialize back to Python
parsed = json.loads(json_str)
print("\nType after loads:", type(parsed))
for person in parsed:
    full = person["first_name"] + " " + person["last_name"]
    print("  ‚Üí", full)
````

#### 6b ¬∑ JSON file serialization ‚Äî `dump()` and `load()`

`dump()` writes JSON to a file-like object and `load()` reads JSON back from files.

````python
import pathlib

out_path = pathlib.Path("people.json")

# Write to file
with open(out_path, "w", encoding="utf-8") as f:
    json.dump(people, f, indent=2)

print("Wrote", out_path.stat().st_size, "bytes to", out_path)

# Read back
with open(out_path, encoding="utf-8") as f:
    reloaded = json.load(f)

print("Reloaded:", [p["first_name"] for p in reloaded])

out_path.unlink()   # tidy up
print("Cleaned up", out_path)
````

## 7 ¬∑ Data Types & Type Introspection

Python is dynamically typed ‚Äî every value carries its type at runtime.
Use `type()` to see the type of a value and `isinstance()` to check membership.

#### 7a ¬∑ Runtime type inspection ‚Äî `type()`

Use `type(value)` to inspect the concrete runtime class of any object.

````python
# ‚îÄ‚îÄ 7a ¬∑ type() ‚Äî runtime type inspection ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Scalar types
first_name = hanu["first_name"]   # str
dob_string = hanu["dob"]          # str
dob_year   = 2002                 # int
height_m   = 1.75                 # float
is_active  = True                 # bool

print(f"first_name : {first_name!r:20s}  ‚Üí {type(first_name)}")
print(f"dob_string : {dob_string!r:20s}  ‚Üí {type(dob_string)}")
print(f"dob_year   : {dob_year!r:20}  ‚Üí {type(dob_year)}")
print(f"height_m   : {height_m!r:20}  ‚Üí {type(height_m)}")
print(f"is_active  : {is_active!r:20}  ‚Üí {type(is_active)}")

# Collection types
person_dict  = hanu                                          # dict
names_list   = [hanu["first_name"], shankar["first_name"]]  # list
hanu_dob     = (20, "Jan", 2002)                             # tuple

print()
print(f"person_dict  : {{...}}               ‚Üí {type(person_dict)}")
print(f"names_list   : {names_list!r:20}  ‚Üí {type(names_list)}")
print(f"hanu_dob     : {hanu_dob!r:20}  ‚Üí {type(hanu_dob)}")
````

#### 7b ¬∑ Type checking ‚Äî `isinstance()`

Use `isinstance(value, T)` (or a tuple of types) for robust type checks that support subclassing.

````python
# isinstance() lets you check a type without knowing it in advance

# Scalar checks
print(f"Is first_name a str?      {isinstance(first_name, str)}")
print(f"Is dob_year an int?       {isinstance(dob_year, int)}")
print(f"Is dob_year a float?      {isinstance(dob_year, float)}")
print(f"Is height_m a float?      {isinstance(height_m, float)}")

# isinstance() also works with tuples of types
print(f"\nIs dob_year int or float?  {isinstance(dob_year, (int, float))}")
print(f"Is height_m int or float?  {isinstance(height_m, (int, float))}")

# Collection checks
print(f"\nIs person_dict a dict?    {isinstance(person_dict, dict)}")
print(f"Is names_list a list?     {isinstance(names_list, list)}")
print(f"Is hanu_dob a tuple?      {isinstance(hanu_dob, tuple)}")

# Useful real-world pattern: accept either list or tuple as a sequence
print(f"\nIs names_list a sequence (list or tuple)?  {isinstance(names_list, (list, tuple))}")
print(f"Is hanu_dob  a sequence (list or tuple)?   {isinstance(hanu_dob,  (list, tuple))}")

# Gotcha: bool is a subclass of int
print(f"\nbool is subclass of int ‚Äî isinstance(True, int): {isinstance(True, int)}")
print(f"type(True) is int: {type(True) is int}")   # False ‚Äî exact match only
````
