## 1 · `for` Loops

`for` iterates over any iterable: lists, dicts, ranges, strings.  
`enumerate()` gives you both the index and value; `zip()` pairs two iterables together.

#### 1a · Basic `for` iteration over a list

A `for` loop visits each element in sequence and executes the same block for each item.

````python
print("── Basic iteration ──")
for person in people:
    full = f"{person['first_name']} {person['last_name']}"
    year = int(person["dob"].split()[-1])
    print(f"  {full} was born in {year}")
````

#### 1b · Indexed iteration with `enumerate()`

`enumerate()` yields `(index, value)` pairs, which is useful for numbered output.

````python
print("── enumerate() ──")
for i, person in enumerate(people, start=1):
    print(f"  {i}. {person['first_name']} {person['last_name']}")
````

#### 1c · Parallel iteration with `zip()`

`zip()` combines multiple iterables so matching positions can be processed together.

````python
first_names = [p["first_name"] for p in people]
last_names  = [p["last_name"] for p in people]

print("── zip() ──")
for fn, ln in zip(first_names, last_names):
    print(f"  {fn} {ln}")
````

#### 1d · Dictionary iteration with `.items()`

Iterating over `.items()` yields key-value pairs for structured field-by-field processing.

````python
print("── Hanu's fields ──")
for key, value in hanu.items():
    print(f"  {key:12}: {value}")
````

## 2 · `while` Loops

`while` keeps looping as long as a condition is `True`.  
Use `break` to exit early, `continue` to skip the rest of an iteration, and the optional `else` clause runs only if the loop exits *without* a `break`.

#### 2a · `while` loop with `break` and `else`

`while` repeats while a condition is true; `else` runs only when the loop finishes without hitting `break`.

````python
test_phones = ["98765432", "hello", "+65 1234 5678", "+44 20 7946 0958"]

i = 0
while i < len(test_phones):
    phone = test_phones[i]
    i += 1
    if phone.startswith("+"):
        print(f"Found international number: {phone}")
        continue
    print(f"  Skipping: {phone!r}")
````

#### 2b · Loop control with `continue`

`continue` skips the rest of the current iteration and advances to the next item.

````python
print("── continue example ──")
entries = ["+65 1234 5678", "", "98765432", None, "+44 20 7946 0958"]
valid = []
for entry in entries:
    if not entry:
        continue
    valid.append(entry)
print("Valid entries:", valid)
````

## 3 · Conditional Blocks

`if / elif / else` selects one branch to execute.  
A **ternary expression** is a one-liner for simple true/false choices.  
Python 3.10+ introduced `match / case` for structural pattern matching.

#### 3a · Branching with `if / elif / else`

Conditional chains evaluate conditions in order and execute the first matching branch.

````python
def classify_phone(phone: str) -> str:
    digits_only = phone.replace("+", "").replace(" ", "")
    if phone.startswith("+"):
        return "international"
    elif digits_only.isdigit() and len(digits_only) == 8:
        return "local"
    else:
        return "unknown"

for person in people:
    result = classify_phone(person["phone"])
    print(f"  {person['first_name']:8} | {person['phone']:16} → {result}")
````

#### 3b · Ternary conditional expressions

A ternary expression provides compact inline branching for simple two-way choices.

````python
for person in people:
    label = "international" if person["phone"].startswith("+") else "local/other"
    print(f"  {person['first_name']}: {label}")
````

#### 3c · Pattern matching with `match / case`

`match` dispatches execution by pattern, offering clearer multi-branch logic than long chains.

````python
for person in people:
    category = classify_phone(person["phone"])
    match category:
        case "international":
            print(f"  {person['first_name']}: ✔ International format")
        case "local":
            print(f"  {person['first_name']}: ℹ Local 8-digit number")
        case _:
            print(f"  {person['first_name']}: ✘ Unknown format")
````

## 4 · Functions

Functions let you name and reuse a block of logic.  
Python supports **positional**, **keyword**, **`*args`** (variable positional), and **`**kwargs`** (variable keyword) parameters.

#### 4a · Function signatures — positional and keyword arguments

Python functions accept positional and keyword arguments, enabling clear and flexible calls.

````python
def full_name(first: str, last: str) -> str:
    return f"{first} {last}"

print(full_name("Hanu", "Madala"))
print(full_name(first="Shankar", last="Cherukuri"))
````

#### 4b · Default parameters and return values

Default arguments provide optional behavior while keeping function calls concise.

````python
from datetime import datetime, date

def calculate_age(dob_str: str, today: date = None) -> int:
    if today is None:
        today = date.today()
    dob = datetime.strptime(dob_str, "%d %b %Y").date()
    return today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))

print(calculate_age("20 Jan 2002"))
print(calculate_age("30 Mar 1997"))
````

#### 4c · Variadic positional arguments — `*args`

`*args` collects extra positional arguments into a tuple for flexible APIs.

````python
def summarise_people(*records):
    for rec in records:
        name = full_name(rec["first_name"], rec["last_name"])
        age = calculate_age(rec["dob"])
        print(f"  {name}, age {age}")

summarise_people(hanu, shankar)
````

#### 4d · Variadic keyword arguments — `**kwargs`

`**kwargs` captures named arguments into a dictionary for extensible parameter sets.

````python
def format_contact(**fields):
    return " | ".join(f"{k}: {v}" for k, v in fields.items())

print(format_contact(name="Hanu Madala", phone="+65 1234 5678", email="hanu.m@tdfs.com"))
````

## 5 · Docstrings & Return Values

A **docstring** is a string literal immediately after the `def` line — it becomes the function's `__doc__` attribute and is shown by `help()`.  
A function can return a single value, a tuple (multiple values), or `None` (implicitly).

#### 5a · Defining docstrings and return contracts

Docstrings document purpose, inputs, and outputs directly on function definitions.

````python
def parse_dob(dob_str: str):
    """
    Parse a DOB string in the format 'DD Mon YYYY'.

    Returns
    -------
    tuple[int, str, int]
        A (day, month_abbr, year) tuple.
    """
    parts = dob_str.split()
    return int(parts[0]), parts[1], int(parts[2])
````

#### 5b · Introspection with `__doc__` and `help()`

Function metadata can be inspected at runtime to improve discoverability and usage.

````python
print(parse_dob.__doc__)
help(parse_dob)
````

#### 5c · Tuple return unpacking

Functions can return multiple values as a tuple, which can be unpacked into separate variables.

````python
day, month, year = parse_dob(hanu["dob"])
print(f"Hanu:    day={day}  month={month}  year={year}")

day, month, year = parse_dob(shankar["dob"])
print(f"Shankar: day={day}  month={month}  year={year}")
````

## 6 · Modularization

As a project grows, splitting code into separate `.py` files makes it easier to maintain and reuse.

**Typical pattern:**
```
notebooks/
    people_utils.py     ← reusable helpers (full_name, calculate_age, …)
    02_control_flow_functions.ipynb
```

In the notebook you would write:
```python
from people_utils import full_name, calculate_age
```

The cell below writes a small helper module to disk and imports from it — demonstrating the same round-trip you would use across a real project.

#### 6a · Defining reusable module code

Moving helper functions into a module improves reuse and keeps notebooks focused on orchestration.

````python
import pathlib

module_code = '''
from datetime import datetime, date

def full_name(first: str, last: str) -> str:
    return f"{first} {last}"

def calculate_age(dob_str: str, today: date = None) -> int:
    if today is None:
        today = date.today()
    dob = datetime.strptime(dob_str, "%d %b %Y").date()
    return today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
'''

pathlib.Path("people_utils.py").write_text(module_code)
print("people_utils.py written.")
````

#### 6b · Importing from a local module

Once written to disk, module symbols can be imported and reused like standard package APIs.

````python
from people_utils import full_name, calculate_age

print("Imported:", full_name.__name__, calculate_age.__name__)
````

#### 6c · Reusing imported helpers in workflows

Imported utilities can be composed in loops and pipelines to keep notebook logic concise.

````python
for person in people:
    print(f"  {full_name(person['first_name'], person['last_name'])}, age {calculate_age(person['dob'])}")
````

## 7 · Lambda Functions, `map()`, `filter()`, `sorted()`

A **lambda** is an anonymous, single-expression function.  
It pairs naturally with the built-ins `map()`, `filter()`, and `sorted()`.

```python
lambda arguments: expression
```

#### 7a · Transformation pipelines with `map()`

`map()` applies a transformation function to each element of an iterable.

````python
full_names = list(map(lambda p: f"{p['first_name']} {p['last_name']}", people))
print("Full names via map():", full_names)
````

#### 7b · Predicate filtering with `filter()`

`filter()` keeps only elements that satisfy a boolean predicate.

````python
intl_only = list(filter(lambda p: p["phone"].startswith("+"), people))
print("International phones:", [p["first_name"] for p in intl_only])
````

#### 7c · Key-based ordering with `sorted()`

`sorted(iterable, key=...)` orders records by a computed key expression.

````python
by_last = sorted(people, key=lambda p: p["last_name"])
print("Sorted by last name:")
for p in by_last:
    print(f"  {p['last_name']}, {p['first_name']}")
````

#### 7d · Datetime-aware sorting with custom keys

Parsing date strings inside the `key` function enables chronological ordering.

````python
from datetime import datetime

by_dob = sorted(
    people,
    key=lambda p: datetime.strptime(p["dob"], "%d %b %Y")
)
print("Sorted by DOB (oldest first):")
for p in by_dob:
    print(f"  {p['first_name']} {p['last_name']} — {p['dob']}")
````
