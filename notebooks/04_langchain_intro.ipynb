# Session 4: LangChain Fundamentals

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableParallel
from src.utils.llm import get_llm

llm = get_llm()
print(f"LLM: {llm.__class__.__name__}")
```

## 1. Basic Chain

```python
prompt = ChatPromptTemplate.from_template(
    "You are a financial analyst. Answer: {question}"
)
chain = prompt | llm | StrOutputParser()
result = chain.invoke({"question": "What are the key risks for tech investors in 2025?"})
print(result[:500])
```

## 2. Parallel Execution with RunnableParallel

```python
fundamental = ChatPromptTemplate.from_template(
    "Analyze fundamentals of {ticker}: revenue growth, margins, P/E ratio"
) | llm | StrOutputParser()

sentiment = ChatPromptTemplate.from_template(
    "Analyze market sentiment for {ticker}: news, analyst ratings, social media"
) | llm | StrOutputParser()

parallel = RunnableParallel({
    "fundamental": fundamental,
    "sentiment": sentiment,
})

results = parallel.invoke({"ticker": "AAPL"})
print("=== Fundamental ===")
print(results["fundamental"][:300])
print("\n=== Sentiment ===")
print(results["sentiment"][:300])
```

## 3. LCEL: LangChain Expression Language

```python
chain = (
    {"query": RunnablePassthrough()}
    | ChatPromptTemplate.from_template("Summarize: {query}")
    | llm
    | StrOutputParser()
)
print(chain.invoke("AAPL Q1 2025 earnings: revenue $120B, EPS $2.50, services at新高"))
```
