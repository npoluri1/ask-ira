## 1. Import Required Libraries

````python
import numpy as np
import pandas as pd

print(f"NumPy  version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
````

## 2. NumPy — Array Creation & Basic Properties

Key factory functions:
- `np.array()` — from Python list/tuple  
- `np.zeros()` / `np.ones()` — filled arrays  
- `np.arange()` — evenly spaced by step  
- `np.linspace()` — evenly spaced by count  

Important attributes: **shape**, **dtype**, **ndim**, **size**

````python
# --- 1D array from a list ---
a1 = np.array([10, 20, 30, 40, 50])
print("1D array:", a1)
print("  shape:", a1.shape, "| dtype:", a1.dtype, "| ndim:", a1.ndim)

# --- 2D array (matrix) ---
a2 = np.array([[1, 2, 3],
               [4, 5, 6]])
print("\n2D array:\n", a2)
print("  shape:", a2.shape, "| size:", a2.size)

# --- Filled arrays ---
print("\nzeros(3,4):\n", np.zeros((3, 4)))
print("\nones(2,3) * 7:\n", np.ones((2, 3)) * 7)

````

## 3. NumPy — Indexing & Slicing

```
arr[start : stop : step]   # slicing syntax
arr[row, col]              # 2D indexing
arr[arr > value]           # boolean mask
```

````python
arr = np.arange(10, 60, 10)          # [10 20 30 40 50]
print("Array:", arr)

# 1D indexing
print("arr[0]:", arr[0])            # first element
print("arr[-1]:", arr[-1])          # last element
print("arr[1:4]:", arr[1:4])        # slice
print("arr[::2]:", arr[::2])        # every other element

# 2D indexing
mat = np.arange(1, 13).reshape(3, 4)
print("\nMatrix:\n", mat)
print("mat[1, 2]:", mat[1, 2])      # row 1, col 2
print("mat[:, 1]:", mat[:, 1])      # entire column 1
print("mat[0:2, 1:3]:\n", mat[0:2, 1:3])  # sub-matrix
````

## 4. NumPy — Mathematical Operations & Statistics

NumPy operations are **element-wise** by default.  
**Broadcasting** lets arrays of different shapes work together without copying data.

````python
x = np.array([1.0, 4.0, 9.0, 16.0, 25.0])
y = np.array([2.0, 2.0, 3.0,  4.0,  5.0])

# Element-wise arithmetic
print("x + y  :", x + y)
print("x * y  :", x * y)
print("x / y  :", x / y)
print("sqrt(x):", np.sqrt(x))

# Broadcasting — add a scalar to every element
print("\nx + 100:", x + 100)

# Aggregation / statistics
grades = np.array([72, 85, 91, 60, 78, 95, 88, 74])
print("\nStudent grades:", grades)
print(f"  Mean  : {np.mean(grades):.2f}")
print(f"  Std   : {np.std(grades):.2f}")
print(f"  Min   : {np.min(grades)}")
print(f"  Max   : {np.max(grades)}")
print(f"  Median: {np.median(grades)}")
````

## 5. Pandas — Series & DataFrame Creation

| Structure | Analogy | Creation |
|-----------|---------|----------|
| `Series` | Labeled 1-D array | `pd.Series(data)` |
| `DataFrame` | Spreadsheet table | `pd.DataFrame(data)` |

````python
# --- Series ---
temperatures = pd.Series([22.5, 19.0, 25.3, 28.1, 17.8],
                         index=["Mon", "Tue", "Wed", "Thu", "Fri"],
                         name="Temperature (°C)")
print("Series:\n", temperatures)
print("\nIndexed value (Wed):", temperatures["Wed"])
````

````python

# --- DataFrame from a dictionary ---
sales_data = {
    "Product":  ["Widget A", "Widget B", "Gadget X", "Gadget Y", "Widget A", "Gadget X"],
    "Region":   ["North", "South", "North", "East", "East", "South"],
    "Units":    [120, 85, 200, 150, 95, 175],
    "Price":    [29.99, 49.95, 99.00, 79.50, 29.99, 99.00],
    "Month":    ["Jan", "Jan", "Jan", "Feb", "Feb", "Feb"],
}
df = pd.DataFrame(sales_data)
df["Revenue"] = df["Units"] * df["Price"]      # derived column

print("\nDataFrame head:")
display(df.head())

print("\nShape:", df.shape)
print("\nColumn dtypes:\n", df.dtypes)
print("\nDescriptive stats:")
display(df.describe())
````

## 6. Pandas — Data Selection & Filtering

| Method | Purpose |
|--------|----------|
| `df["col"]` | Select single column |
| `df[["c1", "c2"]]` | Select multiple columns |
| `df.loc[label]` | Row by **label** |
| `df.iloc[index]` | Row by **integer position** |
| `df[mask]` | Filter rows by boolean condition |

````python
display(df.head())
````

````python
# Column selection
print("Units column:\n", df["Units"].values)
print("\nProduct , Revenue:\n", df[["Product", "Revenue"]])
````

````python

# iloc — by integer position
print("\nFirst 3 rows (iloc):\n", df.iloc[:3])
print("\nRow 2, columns 0-2 (iloc):", df.iloc[2, 0:3].values)
````

````python
# loc — by label (works best with named index)
df_indexed = df.set_index("Product")
print("\nLocate 'Widget A' rows:\n", df_indexed.loc["Widget A"])
````

````python
# Boolean / conditional filtering
high_revenue = df[df["Revenue"] > 10_000]
print("\nHigh-revenue rows (> $10 000):\n", high_revenue)
````

````python

north_widgets = df[(df["Region"] == "North") & (df["Product"].str.startswith("Widget"))]
print("\nNorth-region Widgets:\n", north_widgets)
````

## 7. Pandas — Data Cleaning & Handling Missing Values

Real-world data is messy. Common steps:
1. **Detect** — `isnull()` / `notnull()`
2. **Drop** — `dropna()`
3. **Fill** — `fillna(value_or_strategy)`
4. **Cast** — `astype()`

````python
import math

dirty = pd.DataFrame({
    "Name":   ["Alice", "Bob", None, "Diana", "Eve"],
    "Age":    [25, None, 30, 22, None],
    "Score":  [88.5, 72.0, None, 95.0, 60.0],
    "Grade":  ["A", "C", "B", None, "D"],
})
print("Raw data:")
display(dirty)
````

````python
# Detect missing values
print("\nMissing value count:\n", dirty.isnull().sum())
print("\nAny missing per row:\n", dirty.isnull().any(axis=1).values)
````

````python

# Drop rows with ANY NaN
print("\nAfter dropna():\n", dirty.dropna())
````

````python

# Fill missing values
cleaned = dirty.copy()
cleaned["Age"]   = cleaned["Age"].fillna(cleaned["Age"].median())
cleaned["Score"] = cleaned["Score"].fillna(cleaned["Score"].mean())
cleaned["Name"]  = cleaned["Name"].fillna("Unknown")
cleaned["Grade"] = cleaned["Grade"].fillna("N/A")

print("\nCleaned data:")
display(cleaned)
````

````python

# Type casting
cleaned["Age"] = cleaned["Age"].astype(int)
print("\nAge after astype(int):", cleaned["Age"].tolist())
````

## 8. Pandas — GroupBy & Aggregation

`groupby()` splits a DataFrame into groups, applies a function, and combines the results — the classic **split-apply-combine** pattern.

````python
region_summary = df.groupby("Region").agg(
    Total_Revenue=("Revenue", "sum"),
    Total_Units=("Units", "sum"),
    Num_Orders=("Product", "count"),
)
print("\nRevenue by Region:")
display(region_summary)
````

````python
# Reuse the sales DataFrame from Section 5
print("Original DataFrame:")
display(df)

# Group by Region — total revenue and units per region
region_summary = df.groupby("Region").agg(
    Total_Revenue=("Revenue", "sum"),
    Total_Units=("Units", "sum"),
    Num_Orders=("Product", "count"),
).reset_index()
print("\nRevenue by Region:")
display(region_summary)
````

````python

# Group by Month + Product — total revenue
monthly = df.groupby(["Month", "Product"])["Revenue"].sum().reset_index()
monthly.columns = ["Month", "Product", "Revenue"]
print("\nRevenue by Month & Product:")
display(monthly)
````

````python

# Pivot table — Revenue: rows=Region, columns=Month
pivot = pd.pivot_table(df,
                       values="Revenue",
                       index="Region",
                       columns="Month",
                       aggfunc="sum",
                       fill_value=0)
print("\nPivot Table (Revenue):")
display(pivot)
````

## 9. NumPy ↔ Pandas Integration

Pandas is built on top of NumPy. Conversion between the two is seamless:

| Direction | Method |
|-----------|--------|
| DataFrame → ndarray | `df.values` or `df.to_numpy()` |
| ndarray → DataFrame | `pd.DataFrame(arr, columns=...)` |
| Apply NumPy ufunc on column | `np.log(df["col"])` |

````python
# 1. DataFrame → NumPy
numeric_df = df[["Units", "Price", "Revenue"]]
arr_from_df = numeric_df.to_numpy()
print("DataFrame → ndarray:")
print(arr_from_df)
print("Type:", type(arr_from_df), "| Shape:", arr_from_df.shape)
````

````python

# 2. NumPy → DataFrame
data_np = np.array([
    [101, 5.5, 250.0],
    [102, 6.2, 310.0],
    [103, 4.8, 200.0],
])
df_from_np = pd.DataFrame(data_np, columns=["ID", "Weight_kg", "Value"])
print("\nndarray → DataFrame:")
display(df_from_np)
````

````python

# 3. Apply NumPy ufuncs directly on Pandas columns
df["Log_Revenue"]    = np.log(df["Revenue"])
df["Sqrt_Units"]     = np.sqrt(df["Units"])
df["Revenue_Scaled"] = (df["Revenue"] - df["Revenue"].mean()) / df["Revenue"].std()
print("\nDataFrame with NumPy-derived columns:")
display(df[["Product", "Revenue", "Log_Revenue", "Sqrt_Units", "Revenue_Scaled"]].round(3))
````

````python

# 4. NumPy operations on a column
print("\nCorrelation (np.corrcoef) between Units and Revenue:",
      np.corrcoef(df["Units"], df["Revenue"])[0, 1].round(4))

print("\n--- Demo complete! ---")
````
