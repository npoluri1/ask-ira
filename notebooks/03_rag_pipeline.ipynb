# Session 3b: RAG Pipeline with ChromaDB

```python
from src.rag.vector_store import VectorStore
from src.rag.retrieval import HybridRetriever, rrf_fusion
from src.rag.reranking import CrossEncoderReranker
from src.rag.pipeline import RAGPipeline
from langchain_core.documents import Document

print("RAG modules loaded")
```

## Seed Documents

```python
docs = [
    Document("Apple designs iPhones, Macs, and provides services like App Store and Apple Music.", metadata={"ticker": "AAPL"}),
    Document("Microsoft develops Windows, Office, and Azure cloud platform.", metadata={"ticker": "MSFT"}),
    Document("Apple's services revenue reached $85B in FY2024, up 14% YoY.", metadata={"ticker": "AAPL"}),
    Document("Azure cloud revenue grew 30% YoY, driving Microsoft's growth.", metadata={"ticker": "MSFT"}),
    Document("Both companies face regulatory scrutiny in EU digital markets.", metadata={"topic": "regulation"}),
]
store = VectorStore()
store.add_documents(docs)
print(f"Added {len(docs)} documents")
```

## Hybrid Retrieval (BM25 + Dense)

```python
pipeline = RAGPipeline(store, docs)
results = pipeline.retrieve("What does Apple make?")
print("Retrieved:")
for i, d in enumerate(results, 1):
    print(f"  {i}. {d.page_content}")
```

## Similarity Search

```python
results = store.similarity_search("cloud computing growth")
print("Similarity search results:")
for d in results:
    print(f"  [{d.metadata}] {d.page_content[:80]}...")
```
